```python
import statsmodels.api as sm
from scipy.stats import kstest
import seaborn as sns
from scipy.stats import kstest, norm, expon, gamma, lognorm, weibull_max
import matplotlib.patches as mpatches
```

```python
db_clustered[db_clustered['Cluster']==0].iloc[:,0:1]
```

```python


# Assuming db_clustered is your DataFrame

# Get unique cluster values
cluster_values = db_clustered['Cluster'].unique()

info = [[],[]]

for i in cluster_values:
    info[0].append(i)
    data = db_clustered[db_clustered['Cluster'] == i].iloc[:, 0]  # Adjust column selection

    distributions = [norm, expon, gamma, lognorm, weibull_max]
    params = []

    print(f"\nCluster {i}:")

    for distribution in distributions:
        # Adjust the distribution fit function
        params_dist = distribution.fit(data, loc=0)

        # Calculate the Kolmogorov-Smirnov test
        ks_stat, ks_p_value = kstest(data, distribution.cdf, args=params_dist)

        # Store the results
        params.append({
            'distribution': distribution.name,
            'params': params_dist,
            'ks_stat': ks_stat,
            'ks_p_value': ks_p_value
        })

    # Sort the results based on p-value
    params = sorted(params, key=lambda x: x['ks_stat'])
    info[1].append(params[0]['distribution'])
    for result in params:
        print(f"Distribution: {result['distribution']}, KS Stat: {result['ks_stat']:.4f}, P-value: {result['ks_p_value']:.4f}")

    print(f"The best distribution for Cluster {i} is: {params[0]['distribution']}")

    # Create an individual plot for each cluster
    plt.figure(figsize=(12, 6))

    sns.histplot(data, bins=20, kde=True, color='blue', label=f'Data - Cluster {i}')
    xmin, xmax = plt.xlim()

    x = np.linspace(xmin, xmax, 1000)
    p = getattr(params[0]['distribution'], 'pdf', None)
    if p is not None:
        y = p(x, *params[0]['params'])
        plt.plot(x, y, 'r', linewidth=2, label=f'Best-Fitting Dist. - Cluster {i}')

    plt.title(f'Fit Distribution to Data - Cluster {i}')
    plt.xlabel('Value of Z')
    plt.ylabel('Frequency')
    plt.legend()

    # Save or display the individual plot
    #plt.savefig(f'cluster_{i}_plot.png')
    # Or uncomment the line below to display the plot interactively
    plt.show()

    # Close the current plot to avoid overlapping with the next one
    #plt.close()

```

```python

def get_distribution_values(data, distribution, n=1):
    """
    Get the values of mean plus/minus n standard deviations for a given distribution.

    Parameters:
    - data: array-like, input data
    - distribution: scipy.stats distribution object
    - n: int, number of standard deviations

    Returns:
    - tuple: (lower_bound, mean, upper_bound)
    """
    # Fit the distribution to the data
    params_dist = distribution.fit(data, loc=0)

    # Calculate mean and standard deviation from the fitted distribution parameters
    mean, std = params_dist[-2:]

    # Calculate the lower and upper bounds
    lower_bound = mean - n * std
    upper_bound = mean + n * std

    return lower_bound, mean, upper_bound

# Assuming db_clustered is your DataFrame

# Get unique cluster values
#cluster_values = db_clustered['Cluster'].unique()
results_array = []
# Iterate over clusters
for i, cluster_label in enumerate(info[0]):
    # Subset the DataFrame for the current cluster
    data_cluster = db_clustered[db_clustered['Cluster'] == cluster_label].iloc[:, 0]
    
    
    # Choose the distribution you want to use (you can change this)
    selected_distribution = info[1][i]

    # Get values for mean plus/minus 1 standard deviation
    lower, mean, upper = get_distribution_values(data_cluster, selected_distribution, n=1)

    # Print the results for each cluster
    print(f"\nCluster {cluster_label} - Distribution: {selected_distribution.name}")
    print(f"Mean: {mean:.4f}")
    print(f"Lower Bound: {lower:.4f}, Upper Bound: {upper:.4f}")
    results_array.append([lower, upper])
```
